```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame({
'name':['john','mary','peter','jeff','bill','lisa','jose'],
'age':[23,78,22,19,45,33,20],
'gender':['M','F','M','M','M','F','M'],
'state':['california','dc','california','dc','california','texas','texas'],
'num_children':[2,0,0,3,2,1,4],
'num_pets':[5,1,0,5,2,2,3]
})
df
```

```python
df.plot.scatter(x='gender',y='state',color='blue')
```

```python
df.plot.bar(x='name',y='num_pets')
```

```python
ax = plt.gca()
df.plot.line(x='age',y='num_children',ax=ax)
df.plot.line(x='age',y='num_pets', color='red', ax=ax)
plt.savefig('output2.png')
```

```python
df.plot.bar(x='name',y='age')
# the plot gets saved to 'output.png'
plt.savefig('output.png')
```

```python
df.groupby(['state','gender']).size().unstack().plot.bar(stacked=True)
plt.show()
```
